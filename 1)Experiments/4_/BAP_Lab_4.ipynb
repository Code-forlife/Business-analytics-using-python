```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("multiTimeline (1).csv")
df.columns
```

```python
df["Week"] = pd.to_datetime(df["Week"])
df.set_index("Week", inplace=True)

print("Original Dataset:")
df
```

```python
# Resample to daily frequency (up-sampling with forward fill)
daily_data = df.resample('D').ffill()

# Display the up-sampled daily dataset
print("\nUp-sampled Daily Dataset with Forward Fill:")
print(daily_data)
```

```python
# Resample to monthly frequency (down-sampling with OHLC)
monthly_data = df.resample('M').ohlc()

# Display the down-sampled monthly dataset
print("\nDown-sampled Monthly Dataset with OHLC:")
print(monthly_data)
```

```python
# Shift the time series forward and backward using naive shifts
shifted_forward = df.shift(2)
shifted_backward = df.shift(-2)

# Display the shifted datasets
print("\nShifted Forward Dataset:")
print(shifted_forward)
print("\nShifted Backward Dataset:")
print(shifted_backward)
```

```python
# Shift the time series forward and backward using frequency
shifted_forward_freq = df.shift(2, freq='W')
shifted_backward_freq = df.shift(-2, freq='W')

# Display the frequency-shifted datasets
print("\nFrequency Shifted Forward Dataset:")
print(shifted_forward_freq)
print("\nFrequency Shifted Backward Dataset:")
print(shifted_backward_freq)
```

```python
# Compute moving averages using three different window functions
rolling_mean_3 = df.rolling(window=3).mean()
rolling_sum_4 = df.rolling(window=4).sum()
rolling_expanding = df.expanding().mean()

# Display the computed moving averages
print("\nMoving Average (Window=3):")
print(rolling_mean_3)
print("\nRolling Sum (Window=4):")
print(rolling_sum_4)
print("\nExpanding Mean:")
print(rolling_expanding)
```
